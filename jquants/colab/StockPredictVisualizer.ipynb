```python
import os
import pandas as pd
from google.colab import drive

drive.mount('/content/drive')
refresh_token = ""
with open('/content/drive/MyDrive/drive_ws/jquantsapi-key') as f:
  refresh_token = f.readline().replace('\n', '')

GOOGLE_DRIVE_MOUNT_DIR_PATH = "/content/drive"
RESULT_DIR_PATH = f"{GOOGLE_DRIVE_MOUNT_DIR_PATH}/MyDrive/drive_ws/result/"
predict_result_path = os.path.join(RESULT_DIR_PATH, "predict_result.csv")

def generate_html_table(df):
    # BingSearchUrlとGoogleSearchUrlにHTMLリンクを追加
    df['BingSearchUrl1'] = df['BingSearchUrl1'].apply(lambda x: f'<a href="#" class="clickable-url overflow-text">{x}</a>')
    df['BingSearchUrl2'] = df['BingSearchUrl2'].apply(lambda x: f'<a href="#" class="clickable-url overflow-text">{x}</a>')
    df['GoogleSearchUrl'] = df['GoogleSearchUrl'].apply(lambda x: f'<a href="#" class="clickable-url overflow-text">{x}</a>')
    table_html = df.to_html(classes='table table-bordered', escape=False, index=False, table_id='myTable')

    # 特定のカラムのテキストが途中で切れるようにするスタイルを追加
    style = """
    <style>
    .overflow-text { max-width: 150px; overflow: hidden; text-overflow: ellipsis; white-space: nowrap; display: block; }
    .clickable-url { cursor: pointer; color: #0645ad; text-decoration: underline; }
    </style>
    """

    click_event_js = """
    <script>
    document.querySelectorAll('.clickable-url').forEach(element => {
        element.addEventListener('click', (e) => {
            e.preventDefault();  // デフォルトのアンカータグの挙動を停止
            var url = element.innerText;
            if(url) window.open(url, '_blank');
        });
    });
    </script>
    """
    return HTML(style + table_html + click_event_js)



result: pd.DataFrame = pd.read_csv(predict_result_path, dtype=str)
generate_html_table(result)

#print(result)

```

```python
import pandas as pd
from IPython.display import HTML

# データフレームの作成（例）
data = {
    '名前': ['山田太郎', '鈴木一郎'],
    '年齢': [30, 22],
    'プロフィールURL': ['http://example.com/yamada', 'http://example.com/suzuki']
}
df = pd.DataFrame(data)

# HTMLテーブルの生成とJavaScriptを使ったクリックイベントの追加
def generate_html_table(df):
    table_html = df.to_html(classes='table table-bordered', escape=False, index=False, table_id='myTable')
    click_event_js = """
    <script>
    document.querySelectorAll('#myTable tr').forEach(row => {
        row.addEventListener('click', () => {
            var url = row.cells[row.cells.length - 1].innerText;
            if(url) window.open(url, '_blank');
        });
    });
    </script>
    """
    return HTML(table_html + click_event_js)

# 生成したHTMLテーブルを表示
generate_html_table(df)
```
